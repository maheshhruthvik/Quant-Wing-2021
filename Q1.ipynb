```python
merged [["SMB %" , "HML %" , 'Rf %', 'Rm %', 'Rm-Rf %']] = merged [["SMB %" , "HML %" , 'Rf %', 'Rm %', 'Rm-Rf %']] /100
```

Fama French 3 Factor Model

                        Ri-Rf = Ai + Bi(Rm-Rf) + Si(SMB)+ Hi(HML)+ Ei 
                        

```python
merged.columns = [c.replace(' ', '_') for c in merged.columns]
```

```python
merged['Ri-Rf'] = merged.Returns - merged['Rf_%']
```

```python
merged
```

```python
merged.columns = [c.replace('_', '') for c in merged.columns]

```

```python
merged
```

```python
merged.columns = [c.replace('%', '') for c in merged.columns]
```

```python
y = merged ["Ri-Rf"]
```

```python
x = merged [["Rm-Rf","SMB","HML"]]
```

```python
x_sm= sm.add_constant(x)
```

```python
model = sm.OLS(y,x_sm)
result = model.fit()
result.summary()

```
